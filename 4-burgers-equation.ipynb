```python
from matplotlib import pyplot
%matplotlib inline

###variable declarations
nx = 201
nt = 200
dx = 2 * numpy.pi / (nx - 1)
nu = .07
dt = dx * nu

x = numpy.linspace(0, 2 * numpy.pi, nx)
u0 = numpy.empty(nx)

def Error(x):
    return math.exp(-(x ** 2) / (4 * nu))

def xexp(x):
    return x * math.exp(- (x ** 2) / (4 * nu))

for i in range(nx):
    u0[i] = (xexp(x[i]) + xexp(x[i] - 2. * math.pi)) / (Error(x[i]) + Error(x[i] - 2 * math.pi)) + 4

t = 0

# u = numpy.asarray([ufu0c(t, x0, nu) for x0 in x])
```

```python
pyplot.plot(numpy.linspace(0, 2. * math.pi, nx), u0)
```

```python
u = u0 #initialize a temporary array

for n in range(nt):  #loop for values of n from 0 to nt, so it will ru_old nt times
    u0 = u.copy() ##copy the existing values of u into u_old
    for i in range(1, nx - 1): ## you can try commenting this line and...
    #for i in range(nx): ## ... uncommenting this line and see what happens!
        u[i] = u0[i] + (nu * dt )/ (dx * dx) * (u0[i+1] - 2 * u0[i] + u0[i-1]) - u0[i] * (dt / dx) * (u0[i] - u0[i-1])
    
    u[0] = u0[0] - u0[0] * (dt / dx) * (u0[0] - u0[-2]) + nu * dt / dx**2 * (u0[1] - 2 * u0[0] + u0[-2])
    u[-1] = u[0]        
```

```python
pyplot.plot(numpy.linspace(0, 2 * math.pi, nx), u)
```
